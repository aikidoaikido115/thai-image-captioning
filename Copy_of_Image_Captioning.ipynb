```python
from datasets import load_dataset
from deep_translator import GoogleTranslator
import time

# โหลด dataset
dataset = load_dataset("Naveengo/flickr8k", trust_remote_code=True)


translator = GoogleTranslator(source='en', target='th')


def translate_and_update(batch):

    english_caption = batch["text"]

    translated_caption = translator.translate(english_caption)

    batch["text"] = translated_caption
    return batch

# ขอ 500 พอรัน นานเกิน
translated_dataset = dataset["train"].select(range(500)).map(translate_and_update)


for i in range(10):
    print(f"Original English: {dataset['train'][i]['text']}")
    print(f"Translated: {translated_dataset[i]['text']}\n")
    print(f"Image path: {translated_dataset[i]['image']}")
```

```python
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
```

```python
from datasets import load_dataset
class MyDataset(Dataset):

    def __init__(self):
        # dataset = load_dataset("Naveengo/flickr8k", trust_remote_code=True)
        dataset = translated_dataset
        self.text = []
        for d in dataset:
          self.text.append("image of " + d["text"].lower())
        self.feat = torch.load("flick8k_imagefeatures.pt")

    def __len__(self):
        return 4 # len(self.text)

    def __getitem__(self, idx):
        return {
            "feat": self.feat[idx],
            "text": self.text[idx]
        }
```

```python
train_dataset = MyDataset()
train_dataloader = DataLoader(train_dataset, batch_size=4, shuffle=False) #True)
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer, AutoConfig

class ICModelCross(nn.Module):

  def __init__(self, device):
    super().__init__()

    self.device  = device
    # from CLIP (512) to distillgpt2 (768)
    # write an adapter that maps a CLIP feature vector (dim=512) into a sequence of 10 distillgpt2's feature vectors (dim=768)
    # self.adapter = torch.nn.Linear(512, 10*768).to(self.device)
    self.adapter = torch.nn.Sequential(
        torch.nn.Linear(512, 1024),
        torch.nn.LeakyReLU(),
        torch.nn.Linear(1024, 10*768)
    ).to(self.device)

    config = AutoConfig.from_pretrained("distilgpt2")
    config.add_cross_attention = True

    kwargs_decoder = {}
    kwargs_decoder["config"] = config

    self.decoder = AutoModelForCausalLM.from_pretrained("distilgpt2", **kwargs_decoder, device_map="auto")
    self.tokenizer = AutoTokenizer.from_pretrained("distilgpt2")
    self.tokenizer.add_special_tokens({
        'pad_token': '[PAD]',
        'cls_token': '[CLS]'
    })

    # we still need to adjust the embedding layer
    self.decoder.resize_token_embeddings(len(self.tokenizer))

    # now freeze the decoder, excepted the newly added cross-attention
    # for name, param in self.decoder.named_parameters():
    #   if "cross" in name:
    #     param.requires_grad = True
    #   else:
    #     param.requires_grad = False

    # ignore newly added tokens embeddings for instance
    # # Get the new embedding indices
    # new_token_ids = [self.tokenizer.pad_token_id, self.tokenizer.cls_token_id]
    # new_embeddings = self.decoder.get_input_embeddings().weight.data[new_token_ids]

    # # Unfreeze the new token embeddings
    # for idx in new_token_ids:
    #     self.decoder.get_input_embeddings().weight.data[idx].requires_grad = True

  def generate(self, image_features, prompt):
    image_features = self.adapter(image_features.to(self.device)) # (B, 10*768)
    # print(image)
    image_features = image_features.view((image_features.shape[0], 10, 768)) # (B, 10, 768)

    input = self.tokenizer(prompt, return_tensors="pt", padding=True, truncation=True, max_length=64)

    out = self.decoder.generate(
        input.input_ids.to(self.device),
        encoder_hidden_states=image_features,
        pad_token_id=self.tokenizer.pad_token_id)
    # print(out)
    return prompt + self.tokenizer.batch_decode(out, skip_special_tokens=True)

  def forward(self,
              image_features, # (B, 512)
              text):          # (B, ?)

    image_features = self.adapter(image_features.to(self.device)) # (B, 10*768)
    image_features = image_features.view((image_features.shape[0], 10, 768)) # (B, 10, 768)

    input = self.tokenizer(text, return_tensors="pt", padding=True, truncation=True, max_length=64)

    output = self.decoder(
        input.input_ids.to(self.device),
        encoder_hidden_states=image_features,
        labels=input.input_ids.to(self.device)
    )
    output["labels"] = input.input_ids

    return output
```

## 3.Training loop

```python
mod = ICModelCross("cuda")
# mod = ICModelConcat("cuda")
```

```python
def count_all_parameters(model):
    return sum(p.numel() for p in model.parameters())
```

```python
def count_parameters(model):
    return sum(p.numel() for p in model.parameters() if p.requires_grad)
```

```python
count_parameters(mod), count_all_parameters(mod)
```

```python
# Optimizers specified in the torch.optim package
optimizer = torch.optim.SGD(mod.parameters(), lr=0.001, momentum=0.9)

from tqdm import tqdm

for epoch in range(5_000):

  mod.train(True)
  for t,batch in enumerate(train_dataloader):
      optimizer.zero_grad()

      out = mod(batch["feat"], batch["text"])

      out.loss.backward()
      # print(out.loss.cpu().detach().numpy(), end=" ")

      optimizer.step()
      if (t+1)%100==0:
        print(".", end="", flush=True)

  if epoch%500==0:
    # mod.eval()
    out = mod.generate(batch["feat"][:1], ["image of "])
    print(out)
```
